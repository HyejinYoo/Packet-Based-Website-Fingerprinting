```python
import ast
import numpy as np
```

```python
constant_o = pd.DataFrame()
constant_c = pd.DataFrame()
```

```python
std_dev_out_c = []
std_dev_in_c= []
std_dev_out_o = []
std_dev_in_o= []

avg_out_c = []
avg_out_o = []

flag_close=False

avg_in_c = []
avg_in_o = []

for idx, i in enumerate(features_open['burst']) :
    list = ast.literal_eval(i)
    tmp_o=[]
    tmp_i =[]
    if(y_mul['y'][idx]==-1) :
        flag_close=True
    for j in list:
        if(j==0):
            break
        elif j < 0: #들어오는
            tmp_i.append(j)
        else : #나가는
            tmp_o.append(j)
    std_dev_out_o.append(np.std(np.abs(tmp_o)))
    std_dev_in_o.append(np.std(np.abs(tmp_i)))
    avg_out_o.append(np.mean(tmp_o))
    avg_in_o.append(np.mean(tmp_i))
    if(flag_close==False):
        std_dev_out_c.append(np.std(np.abs(tmp_o)))
        std_dev_in_c.append(np.std(np.abs(tmp_i)))
        avg_out_c.append(np.mean(tmp_o))
        avg_in_c.append(np.mean(tmp_i))
```

```python
flag_close=False

in_pack_num_c = []
out_pack_num_c = []
in_pack_num_o = []
out_pack_num_o = []

in_pack_ratio_c = []
out_pack_ratio_c = []
in_pack_ratio_o = []
out_pack_ratio_o = []

# 10
total_pack_num_c=[]
total_pack_num_o=[]
#6
total_pack_sum_c=[]
total_pack_sum_o=[]
# 8-1 
in_pack_sum_c=[]
in_pack_sum_o=[]
# 8-2
out_pack_sum_c=[]
out_pack_sum_o=[]

for idx, i in enumerate(features_open['pack_size']) :
    list = ast.literal_eval(i)
    if(y_mul['y'][idx]==-1) :
        flag_close=True
    cnt=0
    in_c=0
    out_c=0
    in_s=0
    out_s=0
    for j in list :
        if(j==0) :
            break
        else :
            cnt+=1
            if(j<0) :
                in_s+=j
                in_c+=1
            else :
                out_s +=j
                out_c +=1
    if(flag_close==False) :
        total_pack_num_c.append(cnt)
        in_pack_sum_c.append(in_s)
        out_pack_sum_c.append(out_s)
        total_pack_sum_c.append(in_s+out_s)
        in_pack_num_c.append(in_c)
        out_pack_num_c.append(out_c)
        in_pack_ratio_c.append(in_c/cnt)
        out_pack_ratio_c.append(out_c/cnt)
    total_pack_num_o.append(cnt)
    in_pack_sum_o.append(in_s)
    out_pack_sum_o.append(out_s)
    total_pack_sum_o.append(in_s+out_s)
    in_pack_num_o.append(in_c)
    out_pack_num_o.append(out_c)
    in_pack_ratio_o.append(in_c/cnt)
    out_pack_ratio_o.append(out_c/cnt)
```

```python
#9
total_sum_pack_per_sec_c = []
total_sum_pack_per_sec_o = []

total_num_pack_per_sec_c = []
total_num_pack_per_sec_o = []

flag_close=False

for idx, i in enumerate(features_open['timestamp']) :
    list = ast.literal_eval(i)
    if(y_mul['y'][idx]==-1) :
        flag_close=True
    time_sum= sum(list)
    total_sum_pack_per_sec_o.append(total_pack_sum_o[idx]/time_sum)
    total_num_pack_per_sec_o.append(total_pack_num_o[idx]/time_sum)
    if(flag_close==False) :
        total_sum_pack_per_sec_c.append(total_pack_sum_c[idx]/time_sum)
        total_num_pack_per_sec_c.append(total_pack_num_c[idx]/time_sum)
```

```python
# med 를 이용
#med의 합
sum_med_c=[]
sum_med_o=[]

#med 표준편차
std_dev_med_c = []
std_dev_med_o = []

#med 평균
avg_med_c = []
avg_med_o = []

#burst * med , burst*med avg , burst * med std
sum_burst_med_c = []
sum_burst_med_o = []

avg_burst_med_c = []
avg_burst_med_o = []

std_burst_med_c = []
std_burst_med_o = []

flag_close=False

for idx, i in enumerate(features_open['med']) :
    tmp = []
    burst_med = []
    med = ast.literal_eval(i)
    burst = ast.literal_eval(features_open['burst'][idx])
    burst = [x for x in burst if x != 0]
    med = [x for x in med if x != 0]
    if(y_mul['y'][idx]==-1) :
        flag_close=True
    for it, j in enumerate (med) :
        tmp.append(j)
        burst_med.append(j*it)
            
    sum_med_o.append(sum(tmp))
    std_dev_med_o.append(np.std(np.abs(tmp)))
    avg_med_o.append(np.mean(tmp))
    sum_burst_med_o.append(sum(burst_med))
    avg_burst_med_o.append((np.mean(burst_med)))
    std_burst_med_o.append(np.std(np.abs(burst_med)))
    if (flag_close==False):
        sum_med_c.append(sum(tmp))
        std_dev_med_c.append(np.std(np.abs(tmp)))
        avg_med_c.append(np.mean(tmp))                  
        sum_burst_med_c.append(sum(burst_med))
        avg_burst_med_c.append((np.mean(burst_med)))
        std_burst_med_c.append(np.std(np.abs(burst_med)))
```

```python
#19, 20 처음 30개 incoming, outgoing 갯수, 합 비율
in_num_30_c = []
in_num_30_o = []
out_num_30_c = []
out_num_30_o = []

in_sum_30_c = []
in_sum_30_o = []
out_sum_30_c = []
out_sum_30_o = []

flag_close =False

for idx, i in enumerate(features_open['pack_size']) :
    list = ast.literal_eval(i)
    if(y_mul['y'][idx]==-1) :
        flag_close=True
    in_c=0
    out_c=0
    in_s=0
    out_s=0
    for it,j in enumerate(list) :
        if(j==0 or it>30) :
            break
        else :
            if(j<0) :
                in_c+=1
                in_s+=j
            else :
                out_c +=1
                out_s+=j
    if(flag_close==False) :
        in_num_30_c.append(in_c/30)
        out_num_30_c.append(out_c/30)
        in_sum_30_c.append(in_s/(out_s+in_s))
        out_sum_30_c.append(out_s/(out_s+in_s))
    in_num_30_o.append(in_c/30)
    out_num_30_o.append(out_c/30)
    in_sum_30_o.append(in_s/(out_s+in_s))
    out_sum_30_o.append(out_s/(out_s+in_s))
```
